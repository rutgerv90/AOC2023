```python
import numpy as np
import math
import re
import copy

lineData = []
springOrderData = []


with open("input.txt", "r") as f:
# with open("sample.txt", "r") as f:
    lines = f.read().splitlines()
    for line in lines:
        lsplit = line.split(" ")
        # part1 = [char for char in lsplit[0]]
        part1 = lsplit[0]
        part2 = [int(x) for x in lsplit[1].split(",")]
        lineData.append(part1)
        springOrderData.append(part2)

def calc_possibilities(springs, spaces):
    spaces.reverse()
    springs.reverse()
    # dont forget to reverse item in spaces list
    nrSprings = len(springs)
    nrSpaces = len(spaces)
    diff = nrSprings-nrSpaces
    local_possibilities = []
    if nrSpaces > nrSprings:
        for i,v in enumerate(springs):
            if(v>len(spaces[i])):
                del spaces[i]    
    spaces = spaces[:nrSprings]    
    print(springs, spaces)
    for i,v in enumerate(spaces):
        items = []
        if(i == len(spaces)-1 and i<nrSprings-1):
            items = (springs[-(diff+1):])
        else:
            # print(i,v, springs, spaces)
            items = (springs[i])
        toCheck = list(reversed(spaces[i]))
        # check multiple
        if isinstance(items, list):
            if sum(items)+len(items) == len(toCheck):
                local_possibilities.append(1)
                continue
            else:
                # print(items, toCheck, len(toCheck))
                # make slices
                for i2,v2 in enumerate(items):
                    left = items[:i2]
                    right = items[i2+1:]
                    leftSlice = sum(left)+len(left)
                    rightSlice = sum(right)+len(right)
                    # print(leftSlice, rightSlice)
                    test=  toCheck[leftSlice:len(toCheck)-rightSlice]
                    locs = re.findall(r"(\?*)([\#]+)(\?*)","".join(test))
                    if locs:
                        left2 = locs[0][0]
                        mid2 = locs[0][1]
                        sides = v2-len(mid2)
                        right2 = locs[0][2]
                        # print(left2, mid2, right2, sides, left2[:-sides],right2[:sides])
                        answer = (len(mid2)+len(left2[:-sides])+len(right2[:sides])) +1-v2
                        local_possibilities.append(answer)
                    else:
                        answer = len(test)+1-v2
                        local_possibilities.append(answer)
                        

                    
        else:
            # print('single item')
            answer = len(toCheck)+1-items
            local_possibilities.append(answer)
        
    # print(list(reversed(local_possibilities)))
    # print(math.prod(local_possibilities))
    out = 0
    for i in local_possibilities:
        if i>1:
            out+=i
    if out == 0:
        return 1
    else:
        return out

total = []

# operational (.) or damaged (#)
for i in range(0,len(lineData)):
    # print(lineData[i], springOrderData[i])
    taken = (lineData[i].count('#'))
    openspots = lineData[i].count('?')
    totalneeded = sum(springOrderData[i])
    spaces = re.findall(r"[^\.\n]+",lineData[i])
    result = calc_possibilities(springOrderData[i], spaces)
    total.append(result)

# NOT WORKING PROPERLY YET; NEEDS FIXING
print(sum(total))
    
```
